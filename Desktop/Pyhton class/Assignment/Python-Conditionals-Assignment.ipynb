# Qtn 5

```python
number_1 = int(input("Please Enter Your First Number: "))
number_2 = int(input("Please Enter Your Second Number: "))
operator = input("Please Choose +,-,*,/ : ")

if operator == "+":
    solution = number_1 + number_2
    print(f"Your Solution is {solution}")
elif operator == "-":
    solution = number_1 - number_2
    print(f"Your Solution is {solution}")
elif operator == "*":
    solution = number_1 * number_2
    print(f"Your Solution is {solution}")
elif operator == "/":
    solution = number_1 / number_2
    print(f"Your Solution is {solution}")
else:
    print("Invalid Operation")
```

# Qtn 7

```python
side_1 = int(input("Please Enter the First Side of Your Triangle: "))
side_2 = int(input("Please Enter the Second Side of Your Triangle: "))
side_3 = int(input("Please Enter the Third Side of Your Triangle: "))

if side_1 == side_2 == side_3:
    print(f"All Sides are Equal \nEquilateral Triangle") 
elif any([side_1 == side_2, side_1 == side_3, side_2 == side_3]):
    print(f"Two Sides are Equal \nIsosceles")
elif side_1 != side_2 != side_3:
    print(f"All Sides are Not Equal \nScalene Triangle") 
```
