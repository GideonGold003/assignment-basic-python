# Qtn 5

```python
number = [1,2,2,3,4,5,5,5,6,7,8,9,10]
highest_count = number[0]

for num in number:
    count = number.count(num)
    if count > highest_count:
        highest_count = count
        most_common = num

print("most common: ", most_common)
print("Appears: ", highest_count, "times")
```

# Qtn 6

```python
sentence = input("Please Enter your sentence: ")

vowels = 0
consonants = 0
for letter in sentence.lower():
    if letter in "aeiou":
       vowels += 1
    elif letter >= "a" and letter <= "z":
        consonants += 1

print("Vowels: ", vowels)
print("Consonants: ", consonants)
```

# Qtn 7

```python
sentence = input("Please Enter your sentence: ").split()

longest = sentence[0]

for word in sentence:
    if len(word) > len(longest):
        longest = word
print("Longest word: ", longest)
```
